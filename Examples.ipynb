```python
import squidward.squidward.gpr as gpr
import squidward.squidward.gpc as gpc
import squidward.squidward.kernels as kernels
import squidward.squidward.gp_viz as gp_viz
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1D Regression Problem

```python
samples = 300
x = np.random.uniform(-10,10,samples)
noise = np.random.normal(0,150,samples)
y = (1-x)**3-(1-x)**2+100+noise
```

```python
kernel = kernels.rbf(1000**2,2.0)
model = gpr.gaussian_process(kernel=kernel,var_l=450**2,inv_method='solve')
model.fit(x,y)
x_test = np.linspace(-10,10,100)
mean, var = model.posterior_predict(x_test)
```

```python
plt.figure(figsize=(20,10))
#plt.subplot(1,2,1)
plt.scatter(x,y,alpha=.7,c='r')
plt.xlim(-10,10)
plt.ylim(-1500,1500)
gp_viz.plot_1dr(x_test,mean,var[:,0])
plt.legend()
plt.show()
```

# 2D Regression Case

```python
samples = (500,2)
x = np.random.uniform(-10,10,samples)
noise = np.random.normal(0,1000,samples[0])
y = (1-x[:,0])**3-(1-x[:,0])**2-50*(x[:,1])**2+noise
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.scatter(x[:,0],x[:,1],c=y)
plt.subplot(1,2,2)
plt.scatter(x[:,0],y)
plt.scatter(x[:,1],y)
plt.ylim(-7000,2500)
plt.show()
```

```python
kernel = kernels.rbf2(50**2,10.0)
model = gpr.gaussian_process(kernel=kernel,var_l=500**2,inv_method='inv')
model.fit(x,y)
x_test = np.mgrid[-10:10:1.0,-10:10:1.0].reshape(2,-1).T
mean, var = model.posterior_predict(x_test)
```

```python
gp_viz.plot_2dr(model,True,(-10,10,1))
plt.show()

gp_viz.plot_2dr(model,False)
plt.show()
```

## 2D Classification

```python
n = 100
cov = [[5,0],[0,5]]

x1 = np.random.multivariate_normal(mean=[2,10],cov=cov,size=n)
y1 = np.zeros(n)
x2 = np.random.multivariate_normal(mean=[8,10],cov=cov,size=n)
y2 = np.ones(n)
x3 = np.random.multivariate_normal(mean=[6,3],cov=cov,size=n)
y3 = np.ones(n)*2

x = np.concatenate([x1,x2,x3])
y = np.concatenate([y1,y2,y3]).reshape(-1,1)

X_std = (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0))
x = X_std * (1.0 - 0.0) + 0.0

plt.figure(figsize=(20,5))
plt.scatter(x[:,0],x[:,1],c=y[:,0])
plt.show()
```

```python
kernel = kernels.rbf2(35**2,0.2)
model = gpc.gaussian_process(kernel=kernel,var_l=1.0**2,inv_method='solve')
model.fit(x,y)
x_test = np.mgrid[0:1:0.05,0:1:0.05].reshape(2,-1).T
#x_test = x
mean = model.posterior_predict(x_test)
```

```python
gp_viz.plot_2dc(model,(0,1,0.05))
```

```python
gp_viz.test(model,(-1,2,.1))
plt.scatter(x[:,0],x[:,1],c=y[:,0])
plt.show()
```
